```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import kagglehub
path = kagglehub.dataset_download("jp797498e/twitter-entity-sentiment-analysis")
```

```python
import pandas as pd
df=pd.read_csv(path+"/twitter_training.csv", header=None)
df
```

```python
df.columns = ['tweet_id', 'entity', 'sentiment', 'content']
```

```python
display(df.head())
```

```python
df.drop(columns=['tweet_id'], inplace=True)
df.info()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

LE_sentiment = LabelEncoder()
df['sentiment'] = LE_sentiment.fit_transform(df['sentiment'])

LE_entity = LabelEncoder()
df['entity'] = LE_entity.fit_transform(df['entity'])

df.head()
```

```python
x = df['content']
y = df['sentiment']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=4)
```

```python
tf = TfidfVectorizer(min_df=1,stop_words='english')
x_traincv = tf.fit_transform(x_train)
x_testcv = tf.transform(x_test)
feature_names = tf.get_feature_names_out()
```

```python
x_traincv.toarray()
x_testcv.toarray()
```

```python
mnb = MultinomialNB()
y_train=y_train.astype('int')
y_train
```

```python
mnb.fit(x_traincv,y_train)
```

```python
predictions=mnb.predict(x_testcv)
```

```python
content = input("write a tweet: ")
input_transform = tf.transform([content])
predict = mnb.predict(input_transform)
predicted_sentiment_label = LE_sentiment.inverse_transform(predict)
print(f"tweet state\"{content}\" is: {predicted_sentiment_label[0]}")
```
